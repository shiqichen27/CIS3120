## 1. The CodingGrace Game — Structure Overview

mini-game, ASCII art, and a `player_info` dictionary that tracks state throughout the game.

## 3. Why GitHub for Team Projects?

### Benefits of Version Control

- **Version history** — Track every change to `.py` files. See who changed what and when. Revert to a working version if something breaks.
- **Collaboration** — Multiple people work on the same codebase without overwriting each other. Branches isolate each person's work.
- **Central source of truth** — GitHub is the single authoritative copy of `CodingGraceGame.py`. More reliable than emailing scripts or sharing files.

### Problems Without Version Control

- **Overwriting** — Changes in `CodingGraceGame.py` get overwritten.
- **Lost versions** — No one knows which copy actually works.
- **Debugging** — Hard to isolate who broke a shared function.

## 4. The Branch-Based GitHub Workflow

The six-step workflow for adding a new room:

| Step | Action | Command / Detail |
|---|---|---|
| 1 | Pull latest `main` | `git pull` |
| 2 | Create a feature branch | `git checkout -b dungeon-room` |
| 3 | Implement your room | Add a new room function to `CodingGraceGame.py` following the contract |
| 4 | Test locally | `python CodingGraceGame.py` |
| 5 | Stage, commit, and push | `git add` → `git commit -m "message"` → `git push origin <branch>` |
| 6 | Open a pull request | Create a PR on GitHub; add teammates as reviewers |

The complete command sequence:

```
git pull
→ git checkout -b <branch-name>
→ edit CodingGraceGame.py
→ git add CodingGraceGame.py
→ git commit -m "<descriptive message>"
→ git push origin <branch-name>
```

## 5. The Room Function Contract

Every room function in `CodingGraceGame.py` must follow this contract for consistency
and integration:

1. **Accept** `player_info_arg` as the parameter.
2. **Update the dictionary**: set `location`, modify `health`, optionally add to `inventory` (check for duplicates), append to `choices`.
3. **Call** `show_player_info(player_info_arg)` to display the current state.
4. **Return** either `player_info_arg` (room completes normally) or `"flee"` (player returns to door selection).
5. **May call** `you_died()` or `you_won()` to raise the `GameOver` exception, which `main()` catches.

> **Important:** Do **not** modify `start_new_adventure()`. Only the Integrator edits the
> navigation dispatcher.


### Room Function Template

The cell below provides a complete template that satisfies all five contract requirements.
Study it carefully before writing your own room.


````python
# =============================================================================
# ROOM FUNCTION TEMPLATE
# Copy this cell, rename the function, and customize the narrative and logic.
# =============================================================================

def dungeon_room(player_info_arg):
    """The Dungeon Room: a cold underground chamber with a locked gate puzzle."""

    # Optional: call an ASCII art function here
    print("\nYou have entered the Dungeon Room.")

    # --- Update player state (CONTRACT items 1 and 2) ---
    player_info_arg["location"] = "Dungeon Room"

    damage = 15
    player_info_arg["health"] -= damage

    item = "Rusty Key"
    if item not in player_info_arg["inventory"]:
        player_info_arg["inventory"].append(item)
        print(f"You found a {item} and added it to your inventory!")

    player_info_arg["choices"].append("Dungeon Room")

    # --- Display current state (CONTRACT item 3) ---
    show_player_info(player_info_arg)

    # --- Room narrative and interaction ---
    print("The dungeon is cold and damp. You hear dripping water.")
    print("There is a locked gate to the north and a dark passage to the east.")

    action = input("Do you try the gate or take the passage? > ").lower()

    if "gate" in action:
        if "Rusty Key" in player_info_arg["inventory"]:
            print("The key fits! The gate creaks open.")
            return player_info_arg           # CONTRACT item 4a: normal completion
        else:
            print("The gate is locked and you have no key.")
            return "flee"                     # CONTRACT item 4b: flee
    elif "passage" in action:
        print("The passage leads deeper underground...")
        you_died("The floor collapses beneath you")  # CONTRACT item 5
    else:
        print("You hesitate too long. The walls begin to close in.")
        return "flee"
````

````python
def pink_floral_hall_room(player_info_arg): #CONTRACT ITEM 1
    """The Floral Hall: A Bush Maze filled with water beneath, with a boat to navigate through the corridor."""

    # Call an ASCII art function here
    print("\nYou have entered the Pink Floral Room.")

    # --- Update player state
    player_info_arg["location"] = "Pink Room" #CONTRACT ITEM 2

    player_info_arg["choices"].append("Pink Room") #CONTRACT ITEM 5

    # --- Room narrative and interaction ---
    print("\n--- The Pink Floral Hall ---")
    print("The air is sweet but cold. Beneath a canopy of glowing pink tulips,")
    print("a shallow canal winds through a dense hedge maze.")

    # The Maze
    print("\n--- The Maze or The Alleyway ---")
    while True:
      print("\nChoices: [Boat] through the water maze | Enter the [Alleyway]")
      print("Enter a choice: ", end="")
      action = input(" > ").lower()

      if "boat" in action:
            print("You navigated the boat through the floral hedges of the maze walls.")
            if "Floral Key" not in player_info_arg["inventory"]:
                print("You spot something sparkly underwater... You stick your hand in and found the Floral Key!")
                player_info_arg["inventory"].append("Floral Key") #CONTRACT ITEM 4
                print("The key is now in your inventory!")
            if player_info_arg["health"] < 100: #CONTRACT ITEM 3
              player_info_arg["health"] += 10
              print("The floral scent surrounds you. You gain 10 HP!")

            print("You reach a massive iron gate at the end of the bush water maze.")

            # Choices for the gate
            gate_choice = input("Should you try to open the gate? (yes/no) > ").lower()
            if "y" in gate_choice and "Floral Key" in player_info_arg["inventory"]:
                print("The Floral Key turns smoothly. The gate creaks open to another room!")
                return player_info_arg #CONTRACT ITEM 7
            else:
                print("The gate is locked. You should go back.")
                choice = input("Type [flee] to go back > ").lower()
                if "flee" in choice:
                    continue
                if "flee" not in choice:
                    print("Invalid choice. Please type 'flee'.")
                    continue
                return "flee" #CONTRACT ITEM 7

      elif "alleyway" in action:
            print("The alleyway is pitch black. You feel eyes watching you...")
            # Health penalty is given
            player_info_arg["health"] -= 30 #CONTRACT ITEM 3
            print("A shadowy figure strikes you out of nowhere! You lost 30 HP.")
            if player_info_arg["health"] <= 0:
                return you_died("The shadows claimed you.") # CONTRACT ITEM 8
            else:
                show_player_info(player_info_arg)
                print("You should continue your journey forward.")
                return player_info_arg #CONTRACT ITEM 7

      else:
        invalid_choice = True
        if invalid_choice == True:
          print("Invalid choice. Please choose 'boat' or 'alleyway'.")
          invalid_choice = input("Do you want to try again? (yes/no) > ").lower()
          if "y" in invalid_choice:
            continue

    # --- Display current state
      show_player_info(player_info_arg) #CONTRACT ITEM 6

      return player_info_arg #CONTRACT ITEM 7
````

## 11. Handling Merge Conflicts

### How Conflicts Arise

- Two branches edit the same lines in `CodingGraceGame.py`.
- A room author accidentally edits `start_new_adventure()`.
- Two people modify the same helper function.

Git marks conflicting sections with `<<<<<<<`, `=======`, and `>>>>>>>` markers. The team
must decide what the final version should look like and resolve the markers manually.

### Why the Integrator Role Helps

- Only one person edits the navigation dispatcher.
- Room authors work in isolated sections of the file (their own function definitions).
- This reduces the most common source of conflicts.
- The pattern mirrors real-world development coordination.

## 14. Common Mistakes to Watch For

| Mistake | Why It Matters | How to Avoid |
|---|---|---|
| Committing to `main` | Bypasses review; risks breaking the shared codebase | Always check: `git branch` |
| Forgetting `choices.append()` | Breaks the player history tracking | Follow the contract checklist |
| Duplicate inventory items | Player accumulates repeat items on re-entry | `if item not in inventory` |
| Leaving test calls in code | Breaks the game flow for other team members | Review diff before committing |
| Editing `start_new_adventure()` | Causes merge conflicts with the Integrator | Only Integrator touches this function |
